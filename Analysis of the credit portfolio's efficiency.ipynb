```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 


df = pd.read_csv('cohort_results.csv', sep=';')


df.head()
```

```python
cohort_matrix = df.pivot(index='channel', columns='start_month', values='npl_90_rate_pct')

cohort_matrix
```

```python
plt.figure(figsize=(10, 5))

sns.heatmap(cohort_matrix, annot=True, fmt=".1f", cmap="YlOrRd", cbar_kws={'label': 'NPL 90+ Rate (%)'})

plt.title('Когортный анализ: Доля просрочки 90+ по каналам и месяцам выдачи', fontsize=12, pad=15)
plt.xlabel('Месяц выдачи кредита (start_month)', fontsize=10)
plt.ylabel('Канал привлечения (channel)', fontsize=10)

plt.show()
```
